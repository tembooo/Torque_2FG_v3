```python
import pandas as pd
import matplotlib.pyplot as plt

# üìÇ ÿÆŸàÿßŸÜÿØŸÜ ŸÅÿß€åŸÑ ÿß⁄©ÿ≥ŸÑ
df = pd.read_excel("1.Torque_2FG_v3.xlsx")

# üîπ ÿ≥ÿ™ŸàŸÜ‚ÄåŸáÿß€å F1 ÿ™ÿß F9
feature_cols = [col for col in df.columns if col.startswith("F")]

# üìä ÿß€åÿ¨ÿßÿØ 9 ŸÜŸÖŸàÿØÿßÿ± ÿØÿ± €å⁄© ÿ™ÿµŸà€åÿ± (3 ÿ≥ÿ∑ÿ± √ó 3 ÿ≥ÿ™ŸàŸÜ)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(feature_cols):
    color = 'r' if col == "F9" else 'b'  # ‚úÖ ŸÅŸÇÿ∑ F9 ŸÇÿ±ŸÖÿ≤ ÿ¥ŸàÿØ
    axes[i].plot(df[col], df["Torque"], marker='o', linestyle='-', color=color, label=f"{col} vs Torque")
    axes[i].set_title(f"Torque vs {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Torque")
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()

# üì• ÿ∞ÿÆ€åÿ±Ÿá ÿ™ÿµŸà€åÿ± ÿ®ÿß ŸÜÿßŸÖ ÿØŸÑÿÆŸàÿßŸá
plt.savefig("Torque_vs_Features.png", dpi=300)

plt.show()
```
